## Exercise 6: Fit a linear regression model for life expectancy including a term for an interaction between continent and year. Print out the model in a readable format, e.g., print the coefficients of the model (no need to plot). Hint: adding interaction terms is a form of feature engineering, like we discussed in class (think about, e.g., using (a subset of) polynomial features here).

```python
# One hot encode the continent data (to get Africa in the model)
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder()
enc.fit(data[['continent']])
data[enc.get_feature_names_out()] = pd.DataFrame(enc.transform(data[['continent']]).toarray())
```

```python
import statsmodels.formula.api as sm
import statsmodels.api as sm_1

test = sm_1.add_constant(data[['year', 'continent']])
# using each continent column along with year to create interaction term
model = sm.ols(formula = "lifeExp ~ year * continent_Oceania +  year * continent_Americas +  year * continent_Europe +  year * continent_Africa +  year * continent_Asia", data=data).fit()
model.summary()
```

## Are all parameters in the model significantly (in the p-value sense) different from zero? If not, which are not significantly different from zero? Other libraries (statsmodels or patsy may help you solve this problem)
No, not all the parameters in the model are significantly different from zero; specifically, continent[T.Oceania], year:continent[T.Oceania], continent_Europe, year:continent[T.Europe], continent_Africa, and year:continent_Africa parameters aren't significantly different from zero.

## On average, by how much does life expectancy increase each year for each continent? (Provide code to answer this question by extracting relevant estimates from model fit)
For the Americas, the life expectancy increases each year by 0.1106 years; for Asia, the life expectancy increases each year by  0.1960 years; for Europe, the life expectancy decreases each year by 0.0352 years (although not statistically significant); for Oceania , the life expectancy decreases each year by -0.0468 years (although not statistically significant); for Africa, the life expectancy increases each year by 0.0324 years (although not statistically significant).

## Exercise 7: Make a residuals vs.year violin plot for the interaction model. Comment on how well it matches assumptions of the linear regression model.

```python
fig, ax = plt.subplots()
data["sm_residuals"] = data['lifeExp'] - model.predict(data[enc.get_feature_names_out().tolist() + ['year']])
ax.violinplot(data.groupby(by="year")["sm_residuals"].apply(list), data["year"].unique(),widths=4,showmeans=True)
ax.set_xlabel("Year") 
ax.set_ylabel("Residuals")
ax.set_title("Residuals vs Year") 
```

This matches pretty well the assumptions of the linear regression model as according to the assumptions of the linear regression model, the distribution for each year should be normal and centered at 0. 

# Part 2: Classification
    I chose to use a FIFA22 player dataset from Kaggle, specifically the players_22.csv file (which is found at this url: https://www.kaggle.com/datasets/stefanoleone992/fifa-22-complete-player-dataset?select=players_22.csv). I am trying to predict the overall rating of a player. My predictors are a set of features describing players' skill statistics including power stamina, moving acceleration, etc. I will use the RandomForestRegressor and KNeighborsRegressor algorithms from sklearn. 

```python
# import necessary libraries 
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
# models to use
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
# evaluation metrics to use
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score
# hyperparameters selection procedure
from sklearn.model_selection import GridSearchCV
```

### Obtaining and preparing my data 

```python
# The original dataset
players = pd.read_csv("players_22.csv")
players
```

As seen below, the predictors that will be used are a subset of the 110 features that relate specifically to player statistics:

```python
X = players.select_dtypes(include=np.number).iloc[:, 19:-1]
X
```

The target variable is just the overall column which gives a rating of the player from 0 - 100 (the higher it is, the better the player).

```python
y = players['overall']
y
```

### Preprocessing data
Before training on the data, I need to make sure the input data is normalized so there's no pre-bias towards a certain feature. In addition, I need to impute missing data. In this case, I chose to impute it with the mean of the feature.

```python
# normalizing all data 
for col in X.columns:
    X[col] = (X[col] - X[col].min())/(X[col].max() - X[col].min())

# replacing nans w/ -1 for now (Imputer will be used to replace with mean)
X.fillna(-1, inplace = True)

# Imputing missing data 
imp = SimpleImputer(missing_values = -1, strategy = 'mean')
X = imp.fit_transform(X)

X
```

## Training

I will train using the RandomForestRegressor and KNeighborsRegressor algorithms from sklearn. 

```python
# splitting into train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=33)
```

```python
X_test
```

```python
y
```

### Random Forest Regressor 

For the RandomForestRegressor algorithm, after trying out several for its hyperparameter of the number of estimators, I found that the cross validation scores didn't change considerably after a value as low as 10 estimators. So, I tuned the hyperparameter to 10 estimators.

```python
# Using the GridSearchCV model selection procedure to select best parameters
rf_regressor = RandomForestRegressor(n_estimators = 10)
rf_regressor.fit(X_train, y_train)
```

### K Nearest Regressor 

For the K Nearest Regressor algorithm, I used Grid Search to determine the best value for the hyperparameter of the number of neighbors. It determined that out of the values, 10, 50, 100, 140, and 200, the lowest number of neighbors, 10, worked best:

```python
gridsearch = GridSearchCV(KNeighborsRegressor(), {'n_neighbors':[10, 50, 100, 140, 200]})
gridsearch.fit(X_train, y_train).best_params_
```

```python
knn = KNeighborsRegressor(n_neighbors = 10)
knn = knn.fit(X_train, y_train)
```

## Evaluation + Performance Metrics
To compare the performance of these two models, I use 10 cross-fold validation, mean absolute squared error, and the R^2 coefficient performance metrics.

### 10-fold Cross Validation

```python
# distribution of scores from cross validation
rf_scores = cross_val_score(rf_regressor, X_test, y_test, cv=10)
knn_scores = cross_val_score(knn, X_test, y_test, cv=10)
print("Random Forest Regressor CV scores:", rf_scores)
print("K Nearest Neighbor Regressor CV scores:", knn_scores)
```

```python
# average accuracy rate of cross validation
print("Random Forest Regression Accuracy Rate: ", rf_scores.mean())
print("K Nearest Neighbor Regressor Accuracy Rate: ", knn_scores.mean())
```

```python
# error estimate (standard error)
from math import sqrt
print("Random Forest Regression CV Error: ", rf_scores.std()/sqrt(rf_scores.size))
print("K Nearest Neighbor Regressor CV Error: ", knn_scores.std()/sqrt(knn_scores.size))
```

### Bonus

```python
import scipy.stats as stats
stats.ttest_rel(rf_scores, knn_scores)
```

As seen above, the random forest regression algorithm on average seems to have a 95.15% accuracy while the K nearest neighbor regression algorithm seems to have an accuracy of 92.88% with cross validation. The error estimates of the algorithms, however, seem pretty close with a 0.0018 error estimate for random forest regression and 0.0017 for k nearest neighbor. The difference in these distributions seems statistically significant, as shown by the very low p value of 4.64*10^-7.

### MSE + R^2 Coefficient

```python
# testing the performance of random forest regressor
y_pred = rf_regressor.predict(X_test)
print("Random Forest Regression MSE: ", mean_squared_error(y_test, y_pred))
print("Random Forest R^2 Score: ", r2_score(y_test, y_pred))
# testing the performance of K nearest neighbor regressor
y_pred = knn.predict(X_test)
print("K Nearest Neighbor Regressor MSE:", mean_squared_error(y_test, y_pred))
print("K Nearest Neighbor Regressor R^2 Score: ", r2_score(y_test, y_pred))
```

To see the overall prediction performance, I used MSE and R^2 coefficient. The MSE for Random Forest Regression seems 1 less compared to the MSE for K nearest Neighbor Regressor. This shows that, on average, the values predicted by random forest regressor are closer to the actual values than those predicted by K nearest neighbor. In addition, the R^2 score is about 0.965 for random forest regression compared to 0.94 showing that the random forest regressor accounts for more of the variation.
